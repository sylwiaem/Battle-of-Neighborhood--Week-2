### 2.3 Exploring Areas of Liverpool

#### Creating a function to get all areas of Liverpool.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Town/Area', 
                  'Area Latitude', 
                  'Area Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### 2.4 Analyzing each area

#### In this step we are analyzing each area with venues information.

```python

# one hot encoding
liverpool_onehot = pd.get_dummies(liverpool_venues[['Venue Category']], prefix="", prefix_sep="")

# add area column back to dataframe
liverpool_onehot['Town/Area'] = liverpool_venues['Town/Area'] 

# move area column to the first column
list_column = liverpool_onehot.columns.tolist()
number_column = int(list_column.index('Town/Area'))
list_column = [list_column[number_column]] + list_column[:number_column] + list_column[number_column+1:] 
liverpool_onehot = liverpool_onehot[list_column]

liverpool_onehot.head()
```

#### Rows have been grouped by areas and mean of the frequency of occurences of each category has been calculated.

```python
liverpool_grouped = liverpool_onehot.groupby('Town/Area').mean().reset_index()
liverpool_grouped.head()
```

#### I sorted the venues in descending order and created a new data frame that displays 10 venues of each area.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Town/Area']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
areas_venues_sorted = pd.DataFrame(columns=columns)
areas_venues_sorted['Town/Area'] = liverpool_grouped['Town/Area']

for ind in np.arange(liverpool_grouped.shape[0]):
    areas_venues_sorted.iloc[ind, 1:] = return_most_common_venues(liverpool_grouped.iloc[ind, :], num_top_venues)

areas_venues_sorted.head()
```

### 2.4 Area Clustering

#### For this project I have used KMeans algorithm which is considered as a most common cluster of unsupervised learning. I run K-Means to cluster the areas into 5 clusters due the results I have obtained for k optimum using Elbow Method  

```python
# set number of clusters
kclusters = 5

liverpool_grouped_clustering = liverpool_grouped.drop('Town/Area', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(liverpool_grouped_clustering)

# check cluster labels generated for each row in the dataframe
labels = kmeans.labels_
labels
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=0).fit(liverpool_grouped_clustering)
   
    distortions.append(sum(np.min(cdist(liverpool_grouped_clustering, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / liverpool_grouped_clustering.shape[0])


plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

#### Creating a new dataframe that includes defined clusters as well as the top 10 venues for each area.

### As it seems in above histogram, we can define the ranges as below:
#### 100000 AHP : "Low Level HSP"
#### 100000-150000 AHP : "Mid-1 Level HSP"
#### 150000-200000 AHP : "Mid-2 Level HSP"
#### 200000-250000 AHP : "High-1 Level HSP"
#### '>' 250000 AHP : "High-2 Level HSP"
#### In this case, we can create "Level_labels" with those levels.

```python
level = []
for i in range(0,len(housing_data)):
    if (housing_data['Avg price'][i] < 100000):
        level.append("Low Level HSP")
    elif (housing_data['Avg price'][i] >= 100000 and housing_data['Avg price'][i] < 150000):
        level.append("Mid-1 Level HSP")
    elif (housing_data['Avg price'][i] >= 150000 and housing_data['Avg price'][i] < 200000):
        level.append("Mid-2 Level HSP")
    elif (housing_data['Avg price'][i] >= 200000 and housing_data['Avg price'][i] < 250000):
        level.append("High-1 Level HSP")
    else:
        level.append("High-2 Level HSP")   

housing_data['Level_labels'] = level
housing_data
```

### 3. Summary

#### Meging two tables top 3 table created above with liverpool merged tabel.

```python
import numpy as np

summary = pd.merge(liverpool1_merged, 
                    top3[['Town/Area', 'Top 3 Venues']],
                    left_on = 'Town/Area',
                    right_on = 'Town/Area',
                    how = 'left')
summary= pd.merge(summary, 
                    Cluster_labels[['Clusters', 'Labels']],
                    left_on = 'Cluster Labels',
                    right_on = 'Clusters',
                    how = 'left')
summary = pd.merge(summary, 
                    housing_data[['Town/Area', 'Level_labels']],
                    left_on = 'Town/Area',
                    right_on = 'Town/Area',
                    how = 'left')

summary = summary.drop(['Clusters'], axis=1)
summary.head(3)
```
